```python
#file load and read
import pandas as pd
csvpath1 = r"C:\Users\stave\PyCitySchools\Instructions\PyCitySchools\Resources\schools_complete.csv"
csvpath2 = r"C:\Users\stave\PyCitySchools\Instructions\PyCitySchools\Resources\students_complete.csv"
school_df = pd.read_csv(csvpath1, encoding='utf-8')
student_df = pd.read_csv(csvpath2, encoding= 'utf-8')
school_df
student_df
```

```python
#merge data files into one
complete_df = pd.merge(school_df, student_df, how="left", on=["school_name", "school_name"])
complete_df
```

```python
#determine total number of schools
schoolcount = complete_df["school_name"].nunique()
schoolcount
```

```python
#total number of students in all schools
studentcount = complete_df["Student ID"].nunique()
studentcount
```

```python
#total budget for all schools by school name
budget = complete_df.drop_duplicates(subset=["school_name"])
budget
```

```python
#sum of all school budgets
totalbudget = budget["budget"].sum()
totalbudget
```

```python
#calculate average math score
averagemath = complete_df["math_score"].mean()
averagemath
```

```python
#calculate average reading score
averagereading = complete_df["reading_score"].mean()
averagereading
```

```python
#calculate the percent of students passing math
passmath = complete_df[(complete_df["math_score"] >= 70)].count()["student_name"]
passmathpercent = passmath/float(studentcount) * 100
passmathpercent
```

```python
#calculate the percent of students passing reading
passreading = complete_df[(complete_df["reading_score"] >= 70)].count()["student_name"]
passreadingpercent = passreading/float(studentcount) *100
passreadingpercent
```

```python
#calculate the percent of students passing math and reading
passmathreading = complete_df[(complete_df["math_score"] >= 70) & (complete_df["reading_score"] >= 70)].count()["student_name"]
overallpass = passmathreading /  float(studentcount) * 100
overallpass
```

```python
#create a data frame for the values calculated above with formating
schoolsummary = pd.DataFrame({
    'Total Schools': [schoolcount],
    'Total Students': [studentcount],
    'Total Budget': [totalbudget], 
    'Average Math Score': [averagemath],
    'Average Reading Score': [averagereading], 
    'Percent Passing Math': [passmathpercent], 
    'Percent Passing Reading': [passreadingpercent], 
    'Percent Passing - Overall': [overallpass]
})
schoolsummary["Total Budget"] = schoolsummary["Total Budget"].map("${:,.2f}".format)
schoolsummary["Total Students"] = schoolsummary["Total Students"].map("{:,}".format)
schoolsummary
```

```python
#determine school type
schooltype = school_df.set_index(['school_name'])['type']
schooltype
```

```python
#calculate the number of students per school
perschoolstudentcount = complete_df.groupby(['school_name'])['Student ID'].count()
perschoolstudentcount
```

```python
#calculate individual school budgets
perschoolbudget = complete_df.groupby(["school_name"]).mean()["budget"]
perschoolbudget
```

```python
#calculate the budget per student for each school
percapita = perschoolbudget/perschoolstudentcount.count()
percapita
```

```python
#calculate the average math score for each school
averageschoolmath = complete_df.groupby(['school_name']).mean()['math_score']
averageschoolmath
```

```python
#calculate the average reading score for each school
averageschoolreading = complete_df.groupby(['school_name']).mean()['reading_score']
averageschoolreading
```

```python
#calculate the number of students with passing math score for each school
passmath = complete_df[(complete_df["math_score"] >= 70)]
passmathperschool = passmath.groupby(['school_name']).count()
passmathperschool
```

```python
#calculate the number of students with passing reading score for each school
passreading = complete_df[(complete_df["reading_score"] >= 70)]
passreadingperschool = passreading.groupby(['school_name']).count()
passreadingperschool
```

```python
#calculate the number of students with passing math and reading score for each school
passingmathandreading = complete_df[
    (complete_df["reading_score"] >= 70) & (complete_df["math_score"] >= 70)
]
perschoolpassing = passingmathandreading.groupby(['school_name']).count()
perschoolpassing
```

```python
#calculate percent of students passing math for each school
passmathpercentpassmath = passmath.groupby(['school_name']).count()['student_name']/perschoolstudentcount *100
passmathpercentpassmath
```

```python
#calculate percent of students passing reading for each school
percentpassreading = passreadingperschool['student_name']/perschoolstudentcount *100
percentpassreading
```

```python
#calculate percent of students passing math and reading for each school
schoolpassingpercent = perschoolpassing['student_name']/perschoolstudentcount *100
schoolpassingpercent
```

```python
#create a data frame for all values calculated above with formating
schoolsummary2 = pd.DataFrame({
    'School Type':schooltype,
    'Total Students': perschoolstudentcount, 
    'Total School Budget': perschoolbudget, 
    'Per Student Budget': percapita,
    'Average Math Score': averageschoolmath,
    'Average Reading Score': averageschoolreading,
    '% Passing Math': passmathpercent,
    '% Passing Reading': percentpassreading,
    '% Overall Passing': schoolpassingpercent 
})
schoolsummary2["Total School Budget"] = schoolsummary2["Total School Budget"].map("${:,.2f}".format)
schoolsummary2["Per Student Budget"] = schoolsummary2["Per Student Budget"].map("${:,.2f}".format)
schoolsummary2
```

```python
#determine the top 5 schools by overall passing percent
top5 = schoolsummary2.nlargest(5,'% Overall Passing')
top5
```

```python
#determine the lowest 5 schools by overall passing percent
bottom5 = schoolsummary2.nsmallest(5, '% Overall Passing')
bottom5
```

```python
#create bins for spending per student per school
bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-$630", "$630-$645", "$645-$680"]
spending = schoolsummary2.copy()
spending["Spending per Student"] = pd.cut(schoolsummary2["Per Student Budget"], bins, labels=labels)
spending
```

```python
#calculate math and reading scores and percent passing rates by spending
mathspending = spending.groupby(["Spending per Student"]).mean()["Average Math Score"]
readingpending = spending.groupby(["Spending per Student"]).mean()["Average Reading Score"]
passingmathspending = spending.groupby(["Spending per Student"]).mean()["% Passing Math"]
passingreadingspending = spending.groupby(["Spending per Student"]).mean()["% Passing Reading"]
passingoverallspending = spending.groupby(["Spending per Student"]).mean()["% Overall Passing"]
```

```python
#create a data frame for grades per spending bin
spendingsummary = pd.DataFrame({
    'Average Math Score': mathspending,
    'Average Reading Score': readingpending,
    '% Passing Math': passingmathspending,
    '% Passing Reading': passingreadingspending,
    '% Overall Passing': passingoverallspending 
})
spendingsummary
```

```python
#create bins for school size
bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
size = schoolsummary2.copy()
size["School Size"] = pd.cut(size["Total Students"], bins, labels=labels)
size
```

```python
#create a data frame for grades by school size
sizesummary = pd.DataFrame({
    'Average Math Score': mathsize,
    'Average Reading Score': readingsize,
    '% Passing Math': passingmathssize,
    '% Passing Reading': passingreadingsize,
    '% Overall Passing': passingoverallsize  
})
sizesummary
```
